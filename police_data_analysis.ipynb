```python
import pandas as pd
```

```python
police = pd.read_csv('/content/file (2).csv')
```

```python
police
```

```python
#remove columns that confains missing values
police.isnull().sum()
police.drop(columns = 'country_name', inplace = True)
```

```python
police.head()
#police df will be printed without countr name column
```

```python
#based on filteing and value counts
#for speeding, were men or women stopped more often?
police[police.violation == 'Speeding'].driver_gender.value_counts()
```

```python
#does gender affect during search
police.groupby('driver_gender').search_conducted.sum()
```

```python
police.search_conducted.value_counts()
```

```python
police.stop_duration.value_counts()
```

```python
#mapping & data typecasting
#mean stop_duration
police['stop_duration'] = police['stop_duration'].map({'0-15 Min': 7.5, '16-30 Min': 24, '30+ Min': 45})
police.head(50)
```

```python
police.stop_duration.mean()
```

```python
#groupby & Describe
#compare age distribution for each violation
police.groupby('violation').driver_age.describe()
```
